```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")

print(df.head())
```

a. Compute mean, mode, median, standard deviation, confidence interval, and standard error for each feature.

```python
print("Mean:")
mean = df.mean(numeric_only=True)
print(mean)

print("Median:")
print(df.median(numeric_only=True))

print("Mode:")
print(df.mode(numeric_only=True))

print("Standard Deviation:")
print(df.std(numeric_only=True))

n = len(df)
standard_error = df.std(numeric_only=True) / np.sqrt(n)

print("Standard Error:")
print(standard_error)

lower = mean - 1.96 * standard_error
upper = mean + 1.96 * standard_error

print("Confidence Interval Lower Bound:")
print(lower)

print("Confidence Interval Upper Bound:")
print(upper)
```

b. Compute correlation coefficients between each pair of features and plot a heatmap.

```python
corr = df.corr(numeric_only=True)

print("Correlation:")
print(corr)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True)
plt.show()
```

c. Find covariance between length of sepal and petal.

```python
cov = df["sepal_length"].cov(df["petal_length"])
print("Covariance:", cov)
```
