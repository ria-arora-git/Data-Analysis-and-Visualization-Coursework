```python
import numpy as np
```

Create 1D NumPy array from 10 to 20 (inclusive)

```python
arr = np.arange(10, 21)
print("Original array:", arr)
```

Find sum, mean, and standard deviation

```python
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
```

Convert array to float datatype

```python
arr_float = arr.astype(float)
print("Float array:", arr_float)
```

Multiply every element by 2 and subtract 5

```python
mod_arr = arr_float * 2 - 5
print("After multiply by 2 and subtract 5:", mod_arr)
```

Select all elements greater than 15

```python
print("Elements greater than 15:", arr[arr > 15])
```

Reverse the array using slicing

```python
print("Reversed array:", arr[::-1])
```
