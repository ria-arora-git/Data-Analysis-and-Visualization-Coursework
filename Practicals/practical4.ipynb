```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")

print("Age < 30:", df[df["age"] < 30].shape[0])

print("First class fare total:", df[df["class"]=="First"]["fare"].sum())

print("First class survivors:", df[(df["class"]=="First") & (df["survived"]==1)].shape[0])
print("Second class survivors:", df[(df["class"]=="Second") & (df["survived"]==1)].shape[0])
print("Third class survivors:", df[(df["class"]=="Third") & (df["survived"]==1)].shape[0])
```
