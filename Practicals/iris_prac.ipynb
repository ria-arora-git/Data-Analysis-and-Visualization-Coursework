```python
df = sns.load_dataset('iris')
df.head()
```

```python
print('Numerical:')
len(df.select_dtypes(include=['int64','float64']).columns)
```

```python
print('Categorical:')
len(df.select_dtypes(include=['object','category']).columns)
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, annot= True)
plt.show()
```
