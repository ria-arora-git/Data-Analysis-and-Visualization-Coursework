```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
```

```python
titanic = fetch_openml('titanic', version=1, as_frame=True)
df = titanic.frame
df.head()
```

```python
df = df[['sex','survived','age','fare','pclass']] 
df
```

```python
df = df.dropna()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.show()
```

```python
plt.scatter(df['age'], df['fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.kdeplot(df['age'])
sns.kdeplot(df['fare'])
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```
