```python
import pandas as pd

data = {
    "Name": ["Shah","Vats","Vats","Kumar","Vats","Kumar","Shah","Shah","Kumar","Vats"],
    "Gender": ["Male","Male","Female","Female","Female","Male","Male","Female","Female","Male"],
    "Income": [114000,65000,43150,69500,155000,103000,55000,112400,81030,71900]
}

df = pd.DataFrame(data)
```

a. Family wise total income

```python
print("Shah total:", df[df["Name"]=="Shah"]["Income"].sum())
print("Vats total:", df[df["Name"]=="Vats"]["Income"].sum())
print("Kumar total:", df[df["Name"]=="Kumar"]["Income"].sum())
```

b. Highest and lowest income for each family

```python
print("Shah max:", df[df["Name"]=="Shah"]["Income"].max())
print("Shah min:", df[df["Name"]=="Shah"]["Income"].min())

print("Vats max:", df[df["Name"]=="Vats"]["Income"].max())
print("Vats min:", df[df["Name"]=="Vats"]["Income"].min())

print("Kumar max:", df[df["Name"]=="Kumar"]["Income"].max())
print("Kumar min:", df[df["Name"]=="Kumar"]["Income"].min())
```

c. Income less than 80000

```python
print(df[df["Income"] < 80000])
```

d. Total females and average income

```python
female = df[df["Gender"]=="Female"]
print("Total females:", female.shape[0])
print("Average female income:", female["Income"].mean())
```

e. Delete rows with income less than overall average

```python
avg = df["Income"].mean()
df = df[df["Income"] >= avg]
print(df)
```
