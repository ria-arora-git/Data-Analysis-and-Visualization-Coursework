```python
d1 = pd.Index([1,2,3,4,5])
d2 = pd.Index([5,6])
d3 = pd.Index([1,2,3])
```

```python
print(d1.append(d2))
print(d1.difference(d3))
print(d1.intersection(d3))
print(d1.union(d2))
print(d1.isin([2,3]))
print(d1.delete(1))
print(d1.drop(3))
print(d1.insert(loc=5, item=6))
print(d1.is_monotonic_increasing)
print(d1.is_monotonic_decreasing)
print(d1.is_unique)
print(d1.unique())
```

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
print(obj)
obj2 = obj.reindex(['a', 'b', 'c', 'd', 'e'])
print(obj2)
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3.reindex(range(6), method='ffill')
```

```python
obj3 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj3.reindex(range(6), method='bfill')
```

```python
obj4 = pd.DataFrame(np.arange(9).reshape(3,3), index=['a','b','c'],columns=['Ohio','Texas','California'])
print(obj4)
obj5 = obj4.reindex(['a','b','c','d'],fill_value=2)
print(obj5)
states = ['Ohio','Utah','California']
obj6 = obj4.reindex(columns=states)     #old column data removes column data entirely and add empty new column
print(obj6)
f1 = obj4.drop(['Ohio'],axis=1)
print(f1)
f1 = obj4.drop(['a'])
print(f1)

print(obj4.loc[obj4['Texas'] > 1, ['Ohio', 'Texas']])
```

```python
s=pd.Series([10,20,30], index=['a','b','c'])
print(s)
s1 = s.reindex(['a','b','c','d'], method='ffill')
print(s1)
s2 = s.reindex(['a','b','c','d'], fill_value=1)
print(s2)
s3 = s.reindex(['a','b','c','d','e','f'], method='ffill', limit=2)
print(s3)

s4 = pd.Series([0,1,2], index=[10,20,30])
s5 = s4.reindex([11,21,32], method='nearest',tolerance=2)
print(s5)
```
