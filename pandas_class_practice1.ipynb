```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Asha', 'Ravi', 'Karan', 'Meena', 'Arjun'],
    'Age': [23, 25, 22, 24, 26],
    'Marks': [85, 78, 92, 70, 88],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai']
}

df = pd.DataFrame(data)
df.index = ['a', 'b', 'c', 'd', 'e']
```

```python
print(df)

print(df['Marks'])

print(df[['Name','Age']])

print(df.loc['c'])

print(df.loc['b':'d'])

print("df['Age'] returns a Series")
print("df[['Age']] returns a DataFrame")

print(df.loc['d','Marks'])

print(df.iloc[2,1])

print(df.iloc[0:3,0:2])

print(df.loc['a':'c',['Name','Marks']])

print(".loc slicing is label based and inclusive")
print(".iloc slicing is position based and exclusive of end index")

print(df.at['a','Marks'])
print(df.iat[0,2])

print(df[df['Marks'] > 80])

print(df[df['City'] == 'Mumbai'])

print(df[(df['Age'] >= 23) & (df['Age'] <= 25)])

print(df[(df['City'] == 'Delhi') & (df['Marks'] > 80)])

df_q16 = df.copy()
df_q16.loc[df_q16['Marks'] < 75, 'Marks'] = np.nan
print(df_q16)

df_q17 = df.reindex(['a','b','c','d','e','f'])
print(df_q17)

df_q18 = df.reindex(['e','d','c','b','a'])
print(df_q18)

df_q19 = df.reindex(columns=['Name','Age','Marks','City','Grade'])
print(df_q19)

print("New index labels create rows with NaN values")

df_q21 = df.reindex(['a','b','c','d','e','f'], method='ffill')
print(df_q21)

print(df.drop('c'))

print(df.drop('City', axis=1))

print(df.drop(['a','b']))

print(df.drop('Age', axis=1))

print("axis=0 means rows")
print("axis=1 means columns")
```
