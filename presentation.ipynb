```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [22, 30000, "IT", 0],
    [25, 35000, "HR", 0],
    [30, np.nan, "IT", 0],
    [np.nan, 40000, "Finance", 1],
    [28, 38000, "HR", 0],
    [35, 50000, "IT", 1],
    [40, np.nan, "Finance", 1],
    [29, 42000, "IT", 0],
    [np.nan, 39000, "HR", 1],
    [32, 45000, "IT", 1]
], dtype=object)
```

```python
nums = data[:, :2].astype(float)
cats = data[:, 2]

for i in range(nums.shape[1]):
    nums[np.isnan(nums[:, i]), i] = np.nanmean(nums[:, i])

nums = (nums - nums.min(axis=0)) / (nums.max(axis=0) - nums.min(axis=0))

cats_unique, cats_index = np.unique(cats, return_inverse=True)
cats_onehot = np.eye(len(cats_unique))[cats_index]

X = np.hstack((nums, cats_onehot))
```
