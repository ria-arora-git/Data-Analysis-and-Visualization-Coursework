```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [22, 30000, "IT"],
    [25, 35000, "HR"],
    [30, np.nan, "IT"],
    [np.nan, 40000, "Finance"],
    [28, 38000, "HR"],
    [35, 50000, "IT"],
    [40, np.nan, "Finance"],
    [29, 42000, "IT"],
    [np.nan, 39000, "HR"],
    [32, 45000, "IT"]
], dtype=object)

target = np.array([0, 0, 0, 1, 0, 1, 1, 0, 1, 1])
```

```python
nums = data[:, :2].astype(float)
cats = data[:, 2]

nums[np.isnan(nums)] = np.nanmean(nums, axis=0)
nums = (nums - nums.min(axis=0)) / (nums.max(axis=0) - nums.min(axis=0))

cats_unique, cats_index = np.unique(cats, return_inverse=True)
cats_onehot = np.eye(len(cats_unique))[cats_index]

X = np.hstack((nums, cats_onehot))
```

```python
plt.scatter(nums[:, 0], nums[:, 1])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
