```python
import numpy as np
```

a. Compute the mean, standard deviation, and variance of a two dimensional random integer array along the second axis. 

```python
arr2d = np.random.randint(1, 100, (4, 5))
print("2D Array:\n", arr2d)
print("Mean:", np.mean(arr2d, axis=1))
print("Standard Deviation:", np.std(arr2d, axis=1))
print("Variance:", np.var(arr2d, axis=1))
```

b.  Create a 2-dimensional array of size m x n integer elements, also print the shape, type and data type of the array and then reshape it into an n x m array, where n and m are user inputs given at the run time.

```python
m = int(input("Enter m: "))
n = int(input("Enter n: "))

arr = np.random.randint(1, 50, (m, n))
print("\nOriginal Array:\n", arr)
print("Shape:", arr.shape)
print("Type:", type(arr))
print("Data Type:", arr.dtype)

reshaped_arr = arr.reshape(n, m)
print("Reshaped Array:\n", reshaped_arr)
```

c. Test whether the elements of a given 1D array are zero, non-zero and NaN. Record the indices of these elements in three separate arrays. 

```python
arr1d = np.array([0, 4, np.nan, 7, 0, 9, np.nan])
print("\n1D Array:", arr1d)

zero_indices = np.where(arr1d == 0)
nonzero_indices = np.where(arr1d != 0)
nan_indices = np.where(np.isnan(arr1d))

print("Zero indices:", zero_indices)
print("Non-zero indices:", nonzero_indices)
print("NaN indices:", nan_indices)
```

d. Create three random arrays of the same size: Array1, Array2 and Array3. Subtract Array 2 from Array3 and store in Array4. Create another array Array5 having two times the values in Array1. Find Co variance and Correlation of Array1 with Array4 and Array5 respectively. 

```python
Array1 = np.random.randint(1, 50, (10))
Array2 = np.random.randint(1, 50, (10))
Array3 = np.random.randint(1, 50, (10))

Array4 = Array3 - Array2
Array5 = 2 * Array1

print("\nArray1:", Array1)
print("Array4:", Array4)
print("Array5:", Array5)

print("Covariance:\n", np.cov(Array1, Array4))
print("Correlation:\n", np.corrcoef(Array1, Array5))
```

e. Create two random arrays of the same size 10: Array1, and Array2. Find the sum of the first half of both the arrays and product of the second half of both the arrays. 

```python
Array1 = np.random.randint(1, 10, 10)
Array2 = np.random.randint(1, 10, 10)

print("\nArray1:", Array1)
print("Array2:", Array2)

print("Sum of first half:", np.sum(Array1[:5] + Array2[:5]))
print("Product of second half:", np.prod(Array1[5:] * Array2[5:]))
```
